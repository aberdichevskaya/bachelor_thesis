### word2vec

```python
import nltk
from gensim.models import Word2Vec, LsiModel
from gensim import corpora
import re
```

```python
file = open('../dataset/fit_language_models/text_for_embedding.txt','r')
```

```python
data = file.read()
```

```python
sentences = nltk.sent_tokenize(data)
```

```python
data = []
for s in sentences:
    tokens = nltk.word_tokenize(s)
    data.append(tokens)
```

```python
data
```

```python
model = Word2Vec(data, # главный аргумент - массив с предложениями, предложение - массив со словами
                 window=10, # на каком расстоянии слова будут считаться похожими
                 min_count=1, # будем добавлять слова, которые встретились хотя бы min_count раз
                 workers=4) # запустим параллельно в 4 процесса
```

```python
print(model.wv['врт'])
```

```python
model.wv.save_word2vec_format('../dataset/fit_language_models/word2vec_model.bin', binary=True)
```

## LSA

```python
def create_dict_from_file(file):
    array = {}
    idx = 0
    for l in file:
        k = re.split('\s+', l)
        array[k[0]] = idx
        idx += 1
    return array
```

```python
rus_dict = open("../dictionaries/dictionary.txt", 'r', encoding='utf-8')
rus_dict_array = create_dict_from_file(rus_dict)
```

```python
lsi1m = LsiModel(data, id2word=rus_dict_array, num_topics=200, chunksize=10000, distributed=True)
```
